```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_excel('61331_Classes_FS01-US16_part1.xlsx')
```

```python
df.shape
```

```python
df['course_id'] = df['subj_code'] + df['crse_code']
```

```python
df['term'] = df['term_seq_id'].astype(str) + df['term_code']
```

```python
# drop all summer online courses
df = df[~df.crse_code.str.contains('[A-Z]')]

# drop all LBS courses
df = df[~df.subj_code.str.contains('[LBS]')]
```

```python
df.shape
```

```python
# drop all courses with less than 20 enrollees
course_count = df.groupby('course_id').course_id.count()
courses_geq_20 = course_count[course_count >= 20].index
df = df[df.course_id.isin(courses_geq_20)]
```

```python
courses_taken = df.pivot_table(index='placeholderID', columns='course_id', values='term', aggfunc='first')
```
